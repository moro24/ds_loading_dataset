### Indexing / Slicing / Selecting Sections of the Dataframe 
- Access a Single Value from a row/column pair (.at)
- Access a single value for a row/column pair by integer position. (.iat)
- Access a group of rows and columns by label(s). (.loc)

```python
## getting a single value by the row num and column name 
toyota_df.at[4, 'FuelType']
```

```python
toyota_df.at[9, 'KM']
```

```python
toyota_df.iat[4, 4]
```

```python
toyota_df.iat[9, 3]
```

```python
## all the values in the FuelType column 
toyota_df.loc[15:30, 'FuelType']
```
