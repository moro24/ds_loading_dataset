## Preprocessing of Our Data
- Remove the 'Unamed 0' 

```python
toyota_df.columns
```

```python
toyota_df = toyota_df.drop(columns=['Unnamed: 0'], axis=1)
```

```python
toyota_df
```

```python
## lets convert the KM column to numeric 
toyota_df['KM'] = pd.to_numeric(toyota_df['KM'], errors='coerce')
```

```python
toyota_df.dtypes
```

```python
## lets convert the HP column to numeric 
toyota_df['HP'] = pd.to_numeric(toyota_df['HP'], errors='coerce')
```

```python
toyota_df.dtypes
```

```python
## lets convert the Doors column to numeric 
```

```python
toyota_df
```

```python
## lets replace the value three with 3, four with 4, and five with 5. 
toyota_df['Doors'] = toyota_df['Doors'].replace('three', 3)
toyota_df['Doors'] = toyota_df['Doors'].replace('four', 4)
toyota_df['Doors'] = toyota_df['Doors'].replace('five', 5)
```

```python
## lets convert the HP column to numeric 
toyota_df['Doors'] = pd.to_numeric(toyota_df['Doors'], errors='coerce')
```

```python
toyota_df.dtypes
```

```python
toyota_df.isnull().sum()
```

### Replacing Missing Values 
- Numeric : Mean, Mode, Midean
- Categorical: Group each Cat, Highest Count (Mode)
    - Red = 10, Green = 6, Black = 3 

```python
## cal for average age
mean_age = toyota_df['Age'].mean() 
```

```python
## replace nan with mean_age
toyota_df['Age'] = toyota_df['Age'].fillna(mean_age) 
```

```python
## replacing missing value for KM
toyota_df['KM'] = toyota_df['KM'].fillna(toyota_df['KM'].mean())
```

```python
## replace missing value for FuelType
toyota_df['FuelType'].value_counts()
```

```python
higest_count_fueltype = toyota_df['FuelType'].value_counts().index[0]
higest_count_fueltype
```

```python
toyota_df['FuelType'] = toyota_df['FuelType'].fillna(higest_count_fueltype) 
```

```python
toyota_df.isnull().sum()
```

```python
## replace missing value for HP 
## convert the col to category 
toyota_df['MetColor'] = toyota_df['MetColor'].fillna(toyota_df['MetColor'].value_counts().index[0])
```

```python
##toyota_df['MetColor'].mode()
```

```python
##toyota_df['MetColor'].mean()
```

```python
## replace the missing value with average
toyota_df['HP'] = toyota_df['HP'].fillna(toyota_df['HP'].mean())
```

```python
toyota_df.isnull().sum()
```

```python
## lets convert the MetColor to a category type 
toyota_df['MetColor'] = toyota_df['MetColor'].astype('category')
```

```python
toyota_df.dtypes
```

```python
## lets also convert Automatic , FuelType, and Doors all to category...
toyota_df['Automatic'] = toyota_df['Automatic'].astype('category')
toyota_df['Doors'] = toyota_df['Doors'].astype('category')
toyota_df['FuelType'] = toyota_df['FuelType'].astype('category')
```

```python
toyota_df.dtypes
```
